## Columns in Data

X.1 Internal identification number assigned to each annual contract formal-
ized by an insured.

X.2 Start date of the policyholder’s contract (DD/MM/YYYY).

X.3 Date of last contract renewal (DD/MM/YYYY).

X.4 Date of the next contract renewal (DD/MM/YYYY).

X.5 Date of birth of the insured declared in the policy (DD/MM/YYYY).

X.6 Date of issuance of the insured person’s driver’s license
(DD/MM/YYYY).

X.7 Channel through which the policy was contracted (0: Agent, 1: Insur-
ance brokers).

X.8 Total number of years that the insured has been associated with the
insurance entity.

X.9 Total number of policies held by the insured in the insurance entity.

X.10 Maximum number of policies that the insured has ever had in force.

X.11 Maximum number of products that the insured has simultaneously held
at any given point.

X.12 Number of policies canceled or terminated for nonpayment in the current
year.

X.13 Last payment method of the reference policy (1: half-yearly, 0: annual).

X.14 Net premium amount associated with the policy during the current year.

X.15 Total cost of claims for the insurance policy during the current year.

X.16 Total number of claims incurred for the insurance policy during the
current year.

X.17 Total number of claims filed throughout the entire duration of the policy.

X.18 Ratio of the number of claims filed to the total duration (years) of the
policy in force.

X.19 Type of risk (1: motorbikes, 2: vans, 3: passenger cars, 4: agricultural
vehicles).

X.20 0 for rural, 1 for urban (more than 30,000 inhabitants).

X.21 1 if multiple regular drivers are declared, 0 if only one driver is declared.

X.22 Year of vehicle registration (YYYY).

X.23 Vehicle power measured in horsepower.

X.24 Cylinder capacity of the vehicle.

X.25 Market value of the vehicle as of 31/12/2019.

X.26 Number of vehicle doors.

X.27 Energy source used to power the vehicle (P: Petrol, D: Diesel).

X.28 Vehicle weight in kilograms.

```python
insurance_pre_train['LC'] = insurance_pre_train['X.15'] / insurance_pre_train['X.16']
insurance_pre_train['LC'] = insurance_pre_train['LC'].fillna(0)
insurance_pre_train['HALC'] = insurance_pre_train['LC'] * insurance_pre_train['X.18']

insurance_pre_train['CS'] = (insurance_pre_train['X.16'] > 0).astype(int)
```

```python
insurance_pre_train['X.27'] = insurance_pre_train['X.27'].fillna(insurance_pre_train['X.27'].mode())
```

```python
insurance_pre_train['X.27'].value_counts()
```

```python
insurance_pre_train.drop(columns=['X.1','X.15', 'X.16','X.18', 'X.17'], inplace=True)
```

```python
insurance_pre_train.columns
```

```python
# Make necessary date columns
date_cols = ['X.2','X.3','X.4','X.5','X.6']

# Convert them all to datetime, specifying dayfirst for DD/MM/YYYY
for c in date_cols:
    insurance_pre_train[c] = pd.to_datetime(
        insurance_pre_train[c],
        dayfirst=True,
        errors='coerce'
    )

# Now you can safely compute differences:
insurance_pre_train['policy_duration_days'] = (
    insurance_pre_train['X.4'] - insurance_pre_train['X.2']
).dt.days

insurance_pre_train['time_since_last_renewal_days'] = (
    insurance_pre_train['X.3'] - insurance_pre_train['X.2']
).dt.days

insurance_pre_train['insured_age_years'] = (
    insurance_pre_train['X.2'] - insurance_pre_train['X.5']
).dt.days / 365.25

insurance_pre_train['license_age_years'] = (
    insurance_pre_train['X.2'] - insurance_pre_train['X.6']
).dt.days / 365.25

# Quick check
print(insurance_pre_train[[
    *date_cols,
    'policy_duration_days',
    'time_since_last_renewal_days',
    'insured_age_years',
    'license_age_years'
]].head())
```

```python
insurance_pre_train.drop(columns=['X.2','X.3','X.4','X.5','X.6'], inplace=True)
```

```python
insurance_pre_train.head()
```

```python
insurance_pre_train.isna().sum()
```

```python
# Power-to-Weight Ratio (avoid div by 0)
insurance_pre_train['power_to_weight'] = (
    insurance_pre_train['X.23'] / insurance_pre_train['X.28'].replace(0, np.nan)
)

# Engine Size per Door
insurance_pre_train['engine_per_door'] = (
    insurance_pre_train['X.24'] / insurance_pre_train['X.26'].replace(0, np.nan)
)

# Premium-to-Market Value Ratio
insurance_pre_train['premium_to_value'] = (
    insurance_pre_train['X.14'] / insurance_pre_train['X.25'].replace(0, np.nan)
)
```

```python
# Policy Density
insurance_pre_train['policy_density'] = (
    insurance_pre_train['X.9'] / insurance_pre_train['X.8'].replace(0, np.nan)
)

# Policy Change Ratio
insurance_pre_train['policy_change_ratio'] = (
    insurance_pre_train['X.10'] / insurance_pre_train['X.9'].replace(0, np.nan)
)

# Product Penetration
insurance_pre_train['product_penetration'] = (
    insurance_pre_train['X.11'] / insurance_pre_train['X.9'].replace(0, np.nan)
)

# Cancellation Rate
insurance_pre_train['cancellation_rate'] = (
    insurance_pre_train['X.12'] / insurance_pre_train['X.8'].replace(0, np.nan)
)
```
